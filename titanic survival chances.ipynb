```
# Start 18.30 // 01.01.2023
# End 2.00  /// 02.01.2023

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

```
# Importing Dataset
# 1.1 Please import data set Titanic - Machine Learning from Disaster from https://www.kaggle.com 
df = pd.read_csv('train.csv')
```

```
# 1.2 Realize Exploratory Data Analysis (EDA) by using explanation code 
# https://colab.research.google.com/drive/1CPDcPsg389I_HR6k7WwF1eFEDLaCixZz#scrollTo=iyxqefzIFkc8
#  For Categorical data: https://colab.research.google.com/drive/15GPIyxBdHHkq550GIG19j9vBk-vV4ohs (10 points)
df.corr()
```

```
# My dataset and info about it
print("Train Dataset")
print(df.Survived.value_counts())
print(df.isnull().sum())
df
```

```
# Drop cabin because it is not important and replacing missing values with mean and mode values
df = df.drop(columns='Cabin', axis=1)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```
# Convert gender and Embraked into int values
df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```
# Then select a group of parameters for a model with two or more Independent variables and one 
# Dependent variable=Survived. (10 points)

# Splitting data to X and Y
X = df.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = df['Survived']
```

```
# And splitting X and Y in four variables

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)

print(len(X_train), len(X_test), len(Y_train), len(Y_test))
```

```
# Creating and Train Linear Model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```
x_test_pred = model.predict(X_test) 
r2_score(Y_test, x_test_pred)
```

```
x_train_pred = model.predict(X_train) 
r2_score(Y_train, x_train_pred)
```

```
# Task 2

iris_dataset = load_iris()
print(iris_dataset.keys())
```

```
iris = sns.load_dataset('iris')
iris
```

```
# 2.2 Convert Species columns in a numerical column of the iris dataframe. 
# To encode this data map convert each value to a number. e.g. Iris-setosa:0, 
# Iris-versicolor:1, and Iris-virginica: 2. 
iris.replace({'species':{'setosa':0,'versicolor':1,'virginica':2}}, inplace=True)
iris
```

```
# 2.1 Write a Python program to split the iris dataset into its attributes (X) and labels (y). 
# The X variable contains the first four columns (i.e. attributes) and y contains the labels of the dataset.
# 10 points)

x_train, x_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'],test_size=0.2, random_state=0)
print("x_train: {}, y_train: {} ".format(x_train.shape, y_train.shape[0]))
print("x_test: {},  y_test: {} ".format(x_test.shape,y_test.shape[0]))
```

```
# Now print the iris dataset into 80% train data and 20% test data. Out of a total 150 records,
#  the training set will contain 120 records and the test set contains 30 of those records. Print
#   both datasets. (10 points)/up
x_train
```

```
y_train
```

```
# 2.3 Predict the response for the test dataset (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) 
# using the K Nearest Neighbor Algorithm. Use 5 as the number of neighbors. (10 points)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```
# 2.4 Train or fit the data into the model and calculate the accuracy of the model using the
#  K Nearest Neighbor Algorithm. (10 points)

X_new = np.array([[1, 2, 3, 1]])
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print(": {}".format(
iris_dataset['target_names'][prediction]))
```

```
y_pred = knn.predict(x_test)
print("Prediction for test set:\n {}".format(y_pred))
```

```
print("accuracy: {:.2f}".format(np.mean(y_pred == y_test)))
```

```
# 2.5 Train or fit the data into the model and using the K Nearest Neighbor Algorithm calculate 
# the performance for different values of k (from 1 to 10). (10 points)

knn1 = KNeighborsClassifier(n_neighbors=4)
knn1.fit(x_train, y_train)
```

```
y_pred1 = knn1.predict(x_test)
print("Prediction for test set:\n {}".format(y_pred1))
```

```
print("accuracy: {:.2f}".format(np.mean(y_pred1 == y_test)))
```
